```python
# Read and visualize currency_in_circulation.csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('currency_in_circulation.csv')
df_total = df[df['denomination'] == 'total']
plt.figure(figsize=(80,20))
plt.plot(df_total['date'], df_total['value'], marker='o')
plt.title('RM in Circulation Over Time')
plt.xlabel('Date')
plt.ylabel('Amount (Million RM)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
